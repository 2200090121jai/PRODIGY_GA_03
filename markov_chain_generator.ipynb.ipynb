```python
import random

# 1. Create a larger, custom sample text dataset for the model to learn from
corpus = """
Generative AI is a fascinating field to explore. Generative AI allows us to create text and images.
The journey of learning AI is challenging but rewarding. Every task you complete builds your skills.
Success is built on consistent practice and experimentation. Innovation starts with simple concepts.
Simple concepts grow into complex systems. The future belongs to those who build.
Always keep building, experimenting, and expanding your knowledge in machine learning.
"""

# 2. Function to build the Markov Chain model
def build_markov_chain(text):
    words = text.split()
    markov_chain = {}

    # Loop through words and map which words follow each other
    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        if current_word not in markov_chain:
            markov_chain[current_word] = []
        markov_chain[current_word].append(next_word)

    return markov_chain

# 3. Function to generate new text using a random walk
def generate_text(chain, start_word, num_words=20):
    if start_word not in chain:
        return f"Word '{start_word}' not found in the training text."

    current_word = start_word
    output = [current_word]

    for _ in range(num_words - 1):
        # Look up valid next words from our probability dictionary
        possible_next_words = chain.get(current_word)

        if not possible_next_words:
            break  # Stop if the word has no following states

        # Pick a random next word based on its frequency distribution
        next_word = random.choice(possible_next_words)
        output.append(next_word)
        current_word = next_word

    return " ".join(output)

# 4. Train the model
print("Building the statistical Markov Chain...")
model_chain = build_markov_chain(corpus)
print("Model built successfully!\n")

# 5. Generate text starting with a specific prompt word
starting_prompt = "Generative"
generated_output = generate_text(model_chain, start_word=starting_prompt, num_words=25)

print("--- Statistical Model Generated Text ---")
print(generated_output)
```
